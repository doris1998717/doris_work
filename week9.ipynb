## overfitting

```python
x = np.random.rand(6)
y = np.random.rand(6)
```

```python
plt.scatter(x,y,c="orange",s=100)
```

```python
def myplot(n=1):
    plt.scatter(x,y,c="orange",s=100)
    t = np.linspace(-0.2,1.2,1000)
    plt.plot(t,0.5*np.sin(n*t)+0.5)
myplot()
```

```python
from ipywidgets import interact
```

```python
#interact(套用的函世,參數)
interact(myplot,n=(1,200))
```

真實世界數據不可能長這樣：加上雜訊

$$f(x) = 1.2x + 0.8 + \varepsilon$$

```python
x = np.linspace(0,5,200)
y = 1.2*x + 0.8 + 0.3*np.random.randn(200)
```

```python
plt.scatter(x,y)
```

分訓練資料和測試資料

```python
from sklearn.model_selection import train_test_split
```

### 小重點

機器學習(數據分析)，輸入資料是

$$[x_1,x_2,x_3......]$$
喜歡放成
$$[[x_1],[x_2],[x_3,...]]$$
意思是本來n維向量要變成nx1

```python
x.shape
```

```python
x = x.reshape(200,1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=9487)
#random_state=...為避免每次都重新取樣一次2080%
```

```python
#檢查
len(x_train)
```

```python
len(x_test)
```

## 準備做線性回歸

所有的函數學習，包過線性回歸、機器學習、神經網路都是做三部曲

1.開一台函數學習機
2.訓練
3.預測



### step1：建函數學習機

```python
from sklearn.linear_model import LinearRegression
#linear_model線性模式
```

```python
model = LinearRegression()
#開一台空白的函數學習機
```

## step2：訓練

```python
model.fit(x_train,y_train)
```

## stpe3：預測

```python
##記得輸入格式[]裡的每筆資料也需[]
model.predict([[10]])
```

```python
model.predict([[10],[5],[8]])
```

```python
y_predict = model.predict(x_test)
```

```python
plt.scatter(x_test.ravel(),y_test)
##ravel=拉平
plt.plot(x_test.ravel(),y_predict,"r")
```

波士頓房價預測

```python
from sklearn.datasets import load_boston
```

```python
boston_dataset = load_boston()
```

```python
type(boston_dataset)
```

```python
boston_dataset.feature_names
```

```python
boston_dataset.data[:3]
```

```python
##把boston強制轉換成pd表格(順便把欄位名稱變成feature_names)
boston = pd.DataFrame(boston_dataset.data,columns=boston_dataset.feature_names)
```

```python
boston.head()
```

```python
boston["MEDV"] = boston_dataset.target
```

```python
boston.head()
```

```python
##想看平均房價?
import seaborn as sns
##seaborn：把matplotlib變漂釀
```

```python
sns.distplot(boston.MEDV,bins=30)
```

準備訓練資料

```python
##挑選幾項資料
X = boston.loc[:,"CRIM":"LSTAT"].values
Y = boston.MEDV
```

```python
X[:3]
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=0)
```

## 進入數據分析三部曲

Step1：開函數學習機

```python
model = LinearRegression()
```

Step2：訓練

```python
model.fit(x_train,y_train)
```

Step3：預測

```python
y_predict = model.predict(x_test)
```

```python
##如何得知有沒有預測的好？相減看看
y_predict - y_test
```

```python
##讓圖變漂亮
sns.set()
```

```python
plt.scatter(y_test,y_predict)
plt.xlim(0,55)
plt.ylim(0,55)
plt.plot([0,55],[0,55],"purple")
```
