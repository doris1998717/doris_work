# 105405132 廣告四 宋靜 _ 自製民族風圖案(contourf練習)

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = [1,2,3,5,7,9,13,17,21]
x = np.array(x)
y = [1,2,3,5,7,9,13,17,21]
y = np.array(y)*2
X, Y = np.meshgrid(x, y)

Z = np.sin(X) * np.sin(Y)
plt.contourf(X,Y,Z, cmap="Paired", alpha=0.7)
plt.colorbar()
```
