```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("http://data.coa.gov.tw/Service/OpenData/TransService.aspx?UnitId=QcbUEzN6E6DL&FOTT=CSV")
```

```python
df.head()
```

```python
df2 = df.loc[:,["animal_id","animal_place","animal_kind","animal_sex","animal_bodytype","animal_colour","animal_age","shelter_tel"]]

df2.head()
```

```python
df2.columns = ["ID號碼","收容所","動物類別","性別","體型","毛色","年齡","收容所電話"]
```

```python
df2.head()
```

```python
df3 = df2[(df2.毛色=="黑色") & (df2.動物類別=="狗") & (df2.性別=="F")].head()
df3
```

```python
animal = input("喜歡什麼動物？  ")
flur = input("喜歡什麼毛色？  ")
gender = input("性別？(請回答F或M) ")
size = input("喜歡大型還是小型呢？(請回答SMALL或BIG)  ")

print("\n以下為符合需求隻流浪動物清單，等著你給他們一個幸福的家！\n")
df2[(df2.毛色==str(flur)) & (df2.動物類別==str(animal)) & (df2.性別==str(gender)) & (df2.體型==str(size))]
```
