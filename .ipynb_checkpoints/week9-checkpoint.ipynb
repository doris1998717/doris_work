## overfitting

```python
x = np.random.rand(6)
y = np.random.rand(6)
```

```python
plt.scatter(x,y,c="orange",s=100)
```

```python
def myplot(n=1):
    plt.scatter(x,y,c="orange",s=100)
    t = np.linspace(-0.2,1.2,1000)
    plt.plot(t,0.5*np.sin(n*t)+0.5)
myplot()
```

```python
from ipywidgets import interact
```

```python
#interact(套用的函世,參數)
interact(myplot,n=(1,200))
```

```python
x = np.linspace(0,5,200)
y = 1.2*x + 0.8 + 0.3*np.random.randn(200)
```
