# 發票對對對

```python
prizes = []
pri6s = []

priSpe = input("特別獎號碼為？  ")
pri = input("特獎號碼為？  ")
for i in range(3):
    pri1 = input("頭獎號碼{}為？  ".format(i+1))
    prizes.append(pri1)
for i in range(2):
    pri2 = input("增開六獎號碼{}為？  ".format(i+1))
    pri6s.append(pri2)

print('【開始對獎】')
while True:
    num = input("輸入發票後三碼(q離開)：  ")

    if num == 'q':
        break

    elif num == priSpe[-3:]:
        full = input("有機會喔!請輸入完整號碼：")
        if full == priSpe:
            print("天哪!!!!!!你中了特別獎1000萬元。")
        else:
            print("可惜沒中，繼續加油!")
            
    elif num == pri[-3:]:
        full = input("有機會喔!請輸入完整號碼：")
        if full == pri:
            print("天哪!!!!!!你中了特獎200萬元。")
        else:
            print("可惜沒中，繼續加油!")

    elif num in [i[-3:] for i in prizes]:
        p = prizes[[i[-3:] for i in prizes].index(num)]
        full = input("中獎了!請輸入完整號碼：")
        if full == p:
            print("天哪!!!!!!你中了頭獎20萬元。")
        elif full[-7:] == p[-7:]:
            print("天哪!!!!!!你中了貳獎4萬元。")
        elif full[-6:] == p[-6:]:
            print("天哪!!!!!!你中了參獎1萬元。")
        elif full[-5:] == p[-5:]:
            print("天哪!!!!!!你中了肆獎4000元。")
        elif full[-4:] == p[-4:]:
            print("天哪!!!!!!你中了伍獎1000元。")
        elif full[-3:] == p[-3:]:
            print("天哪!!!!!!你中了陸獎200元。") 
    
    elif num in pri6s:
        print("天哪!!!!!!你中了增開陸獎200元。")
    
    else:
        print('嗚嗚..沒有中喔!')

```

# 我們與愛的距離

```python
import random

print("他")
h = random.randint(2,60)
for i in range(1,h):
    if i != h-1:
        print('♥'*i)
    else:
        print('♥'*i+"你")
```

# 健身幫你配!

```python
import random
workout = {"仰臥起坐":7.17, "伏地挺身":6, "跳繩":8.4, "慢跑":4.7}

burn = int(input("今天想消耗多少熱量?(kcal) "))
w = random.choice(list(workout.keys()))
result = round(burn / workout[w])
print("哇!感覺今天可以" + w + str(result) + "分鐘唷")
```

# 蓋沙漏

```python
res = ''
size = int(input('要蓋個幾層好呢?  '))
lines = list(range(1, int(size) + 1))[::-2] + list(range(3, int(size) + 1))[::2]

for line in lines:
    res += int((size-line)/2)*' ' + line*'*'+'\n'

print(res)
```
